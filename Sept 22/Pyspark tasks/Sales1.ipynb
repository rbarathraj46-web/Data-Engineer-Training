```python
import pandas as pd
import numpy as np
data = {
    "Product" : ["Tv", "Frigde", "WM", "AC", "Oven"],
    "Sales" : [4,3,5,6,3],
    "Price" : [200000,300000,250000,350000,100000]
}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
avg_sales = df.groupby("Product")["Price"].mean()
avg_sales.plot(kind = "bar", color = "red", edgecolor = "black")
plt.title("Average Sales by price - Barath")
plt.xlabel("Product")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```
