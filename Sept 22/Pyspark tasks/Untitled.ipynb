```python
import pandas as pd
import numpy as np
data = {
    "Name" : ["Rahul", "priya", "arjun", "neha", "vikram"],
    "Age" : [21,22,20,23,21],
    "Course" :["AI", "ML", "DATA", "AI", "ML"],
    "Marks" : [85,90,78,88,95]
}
df = pd.DataFrame(data)
print(df)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
avg_marks = df.groupby("Course")["Marks"].mean()
avg_marks.plot(kind = "bar", color = "red", edgecolor = "black")
plt.title("Average marks by course - Barath")
plt.xlabel("Course")
plt.ylabel("Average Marks")
plt.xticks(rotation=45)
plt.show()
```
