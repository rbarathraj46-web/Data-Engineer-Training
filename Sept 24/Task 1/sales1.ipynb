```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Revenue": [12000, 15000, 17000, 14000, 18000, 22000,
                24000, 20000, 23000, 25000, 27000, 30000],
    "Profit": [3000, 3500, 4000, 3200, 4500, 6000,
               6500, 5800, 6200, 7000, 7500, 8000],
    "Region": ["North", "South", "East", "West", "North", "South",
               "East", "West", "North", "South", "East", "West"]
}

df = pd.DataFrame(data)
print("First 5 rows of dataset:")
print(df.head())
```

```python
plt.figure(figsize=(10,6))
plt.plot(df["Month"], df["Revenue"], marker='o', linestyle='-', color='b')
plt.title("Revenue over Months")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Month", y="Profit", data=df, palette="viridis")
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(10,6))
bar_width = 0.35
months = df["Month"]
plt.bar(df.index - bar_width/2, df["Revenue"], bar_width, label="Revenue")
plt.bar(df.index + bar_width/2, df["Profit"], bar_width, label="Profit")
plt.xticks(df.index, months)
plt.title("Revenue vs Profit per Month")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,8))
region_revenue = df.groupby("Region")["Revenue"].sum()
plt.pie(region_revenue, labels=region_revenue.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set2"))
plt.title("Revenue Contribution by Region")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="Revenue", y="Profit", data=df, hue="Region", s=100)
plt.title("Revenue vs Profit (Correlation)")
plt.xlabel("Revenue")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df["Revenue"], bins=8, kde=True, color="blue")
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Region", y="Profit", data=df, hue="Region", palette="Set3")
plt.title("Profit across Regions")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```
